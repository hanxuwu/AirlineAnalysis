# Airline Data Generator

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bz2
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)

#Filenames declaration
fsearches='searches_sample.csv.bz2'
fbookings='bookings_sample.csv.bz2'

#searches_df=pd.read_csv(fsearches,sep='^')
bookings_df=pd.read_csv(fbookings,sep='^')

#big_searches_df=searches_df.copy()
big_bookings_df=bookings_df.copy()
#print("origin len",len(searches_df))
#searches_df.head(5)
```

###  Duplicate *2000

searches sample

```python
for j in range(2000):
    big_searches_df=pd.concat([big_searches_df,searches_df])
    if j%100==0:
        print(j)
        
        print(len(big_searches_df))
print('done')
```

###  Save to CSV

```python
import csv
big_searches_df.to_csv("big_searches.csv",quoting=csv.QUOTE_NONNUMERIC,index=False,sep='^')
```
