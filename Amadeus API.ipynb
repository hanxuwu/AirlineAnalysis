## Flight Most Booked Destinations

provides a list of the most popular flight destinations from a given origin during a specific time period. The popularity is a measure of the total number of bookings made for that destination from that origin (PNRs created).

This answers questions like "Where did most people book their flights to from Paris in September 2016?"

The response is a list of air-traffic objects with the following information:

* destination city

* flights score: ratio of the number of flights to the destination from that origin and the total number of flights from that origin (the ratio will be a number between 0 and 1, 1 being the score for the top destination)

* travelers score: ratio of the number of passengers traveling to the destination from that origin and the total number of passengers traveling from that origin (the ratio will be between 0 and 1, 1 being the score for the top destination). The list is ordered by travelers score in descending order by default. Sorting by flights score is also allowed.

### What were the top destinations people booked flights to from SYD in August 2017?

```python
response_best_place = amadeus.travel.analytics.air_traffic.traveled.get(originCityCode='SYD', period='2017-08')
response_best_place = response_best_place.data
#print(response_best_place)
response_best_place_df=pd.DataFrame.from_dict(json_normalize(response_best_place), orient='columns')
response_best_place_df
```

### Adding the region

```python
# Get a specific city or airport based on its id

airport_codes = response_best_place_df['destination'].tolist()
# airport_codes

test_list = []

for airport in airport_codes:
    try:
        response = amadeus.reference_data.locations.get(
        keyword=airport,
        subType=Location.ANY)

     #  print(airport) 

        test_list.append(response.data[0]['address']['regionCode'])

    except:
        test_list.append('not here')
    

response_best_place_df['country'] = pd.DataFrame(test_list)
response_best_place_df
```
